```python
# Derivadas parciales para regresión lineal
# ####################################################################
def der_parcial_1(xs: list, ys: list) -> tuple[float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 1 al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
        c_1 * a_1 + c_0 * a_0 = c_ind

    ## Parameters
        # ``xs``: lista de valores de x.
        ``ys``: lista de valores de y.
        ## Return
        # ``c_1``: coeficiente del parámetro 1.
        ``c_0``: coeficiente del parámetro 0.
        ``c_ind``: coeficiente del término independiente.
    """
    # coeficiente del término independiente
    c_ind = sum(ys)
    # coeficiente del parámetro 1
    c_1 = sum(xs)
    # coeficiente del parámetro 0
    c_0 = len(xs)
    return (c_1, c_0, c_ind)


def der_parcial_0(xs: list, ys: list) -> tuple[float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro 0 al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
        c_1 * a_1 + c_0 * a_0 = c_ind

    ## Parameters
        # ``xs``: lista de valores de x.
        ``ys``: lista de valores de y.
        ## Return
        # ``c_1``: coeficiente del parámetro 1.
        ``c_0``: coeficiente del parámetro 0.
        ``c_ind``: coeficiente del término independiente.
    """
    c_1 = 0
    c_0 = 0
    c_ind = 0
    for xi, yi in zip(xs, ys):
        # coeficiente del término independiente
        c_ind += xi * yi
        # coeficiente del parámetro 1
        c_1 += xi * xi
        # coeficiente del parámetro 0
        c_0 += xi
        return (c_1, c_0, c_ind)
```

# Conjunto de datos 2 - Interpolación cúbica
Se elige interpolar usando una ecuación cúbica de la forma:
$\hat{y}(x) = a x^3 + b x^2 + c x + d$

La función de pérdida $\mathcal{L}$ se define como el error cuadrático medio entre los valores predichos $\hat{y}$ y los valores reales $y$:
$\mathcal{L} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$


$\nabla \mathcal{L} = [\frac{\partial \mathcal{L}}{\partial a}, \frac{\partial \mathcal{L}}{\partial b}, \frac{\partial \mathcal{L}}{\partial c}, \frac{\partial \mathcal{L}}{\partial d}]=0$

El sistema de ecuaciones resultante es:
\begin{align}
\frac{\partial \mathcal{L}}{\partial a} \rightarrow \sum_{i=1}^{n} (y_i - (a x_i^3 + b x_i^2 + c x_i + d)) x_i^3 = 0 \\
\frac{\partial \mathcal{L}}{\partial b} \rightarrow \sum_{i=1}^{n} (y_i - (a x_i^3 + b x_i^2 + c x_i + d)) x_i^2 = 0 \\
\frac{\partial \mathcal{L}}{\partial c} \rightarrow \sum_{i=1}^{n} (y_i - (a x_i^3 + b x_i^2 + c x_i + d)) x_i = 0 \\
\frac{\partial \mathcal{L}}{\partial d} \rightarrow \sum_{i=1}^{n} (y_i - (a x_i^3 + b x_i^2 + c x_i + d)) = 0
\end{align}


```python
# Derivadas parciales para regresión cúbica
# ####################################################################
def der_parcial_a(xs: list, ys: list) -> tuple[float, float, float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro ``a`` al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
        c_1 * a + c_2 * b + c_3*c + c_4*d = c_ind

    ## Parameters
        ``xs``: lista de valores de x.
        ``ys``: lista de valores de y.
        ## Return
        ``c_1``: coeficiente del parámetro a.
        ``c_2``: coeficiente del parámetro b.
        ``c_3``: coeficiente del parámetro c.
        ``c_4``: coeficiente del parámetro d.
        ``c_ind``: coeficiente del término independiente.
    """
    c_ind = 0
    c_1 = 0
    c_2 = 0
    c_3 = 0
    c_4 = 0
    for xi, yi in zip(xs, ys):
        c_ind += yi * (xi**3)
        c_1 += xi**6
        c_2 += xi**5
        c_3 += xi**4
        c_4 += xi**3
    return (c_1, c_2, c_3, c_4, c_ind)


def der_parcial_b(xs: list, ys: list) -> tuple[float, float, float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro ``b`` al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
        c_1 * a + c_2 * b + c_3*c + c_4*d = c_ind

    ## Parameters
        ``xs``: lista de valores de x.
        ``ys``: lista de valores de y.
        ## Return
        ``c_1``: coeficiente del parámetro a.
        ``c_2``: coeficiente del parámetro b.
        ``c_3``: coeficiente del parámetro c.
        ``c_4``: coeficiente del parámetro d.
        ``c_ind``: coeficiente del término independiente.
    """
    c_ind = 0
    c_1 = 0
    c_2 = 0
    c_3 = 0
    c_4 = 0
    for xi, yi in zip(xs, ys):
        c_ind += yi * (xi**2)
        c_1 += xi**5
        c_2 += xi**4
        c_3 += xi**3
        c_4 += xi**2
    return (c_1, c_2, c_3, c_4, c_ind)


def der_parcial_c(xs: list, ys: list) -> tuple[float, float, float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro ``c`` al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
        c_1 * a + c_2 * b + c_3*c + c_4*d = c_ind

    ## Parameters
        ``xs``: lista de valores de x.
        ``ys``: lista de valores de y.
        ## Return
        ``c_1``: coeficiente del parámetro a.
        ``c_2``: coeficiente del parámetro b.
        ``c_3``: coeficiente del parámetro c.
        ``c_4``: coeficiente del parámetro d.
        ``c_ind``: coeficiente del término independiente.
    """
    c_ind = 0
    c_1 = 0
    c_2 = 0
    c_3 = 0
    c_4 = 0
    for xi, yi in zip(xs, ys):
        c_ind += yi * xi
        c_1 += xi**4
        c_2 += xi**3
        c_3 += xi**2
        c_4 += xi
    return (c_1, c_2, c_3, c_4, c_ind)


def der_parcial_d(xs: list, ys: list) -> tuple[float, float, float, float, float]:
    """Retorna los coeficientes de la ecuación de la derivada parcial con respecto al parámetro ``d`` al reemplazar los valores ``xs`` y ``ys``. La ecuación es de la forma:
        c_1 * a + c_2 * b + c_3*c + c_4*d = c_ind

    ## Parameters
        ``xs``: lista de valores de x.
        ``ys``: lista de valores de y.
        ## Return
        ``c_1``: coeficiente del parámetro a.
        ``c_2``: coeficiente del parámetro b.
        ``c_3``: coeficiente del parámetro c.
        ``c_4``: coeficiente del parámetro d.
        ``c_ind``: coeficiente del término independiente.
    """
    c_ind = 0
    c_1 = 0
    c_2 = 0
    c_3 = 0
    c_4 = 0
    for xi, yi in zip(xs, ys):
        c_ind += yi
        c_1 += xi**3
        c_2 += xi**2
        c_3 += xi
        c_4 += 1
    return (c_1, c_2, c_3, c_4, c_ind)
```

```python
a, b, c, d = ajustar_min_cuadrados(
    xs, ys, gradiente=[der_parcial_a, der_parcial_b, der_parcial_c, der_parcial_d]
)
print(f"a: {a}, b: {b}, c: {c}, d: {d}")
```
